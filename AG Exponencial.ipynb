```python
import numpy as np
import random

print("Digite seu modelo:")
input_model = input()
modelo = [int(i) for i in input_model.split()]
print("\n")
print("Modelo: {}".format(modelo))
print("\n")
tamCromossomo=len(modelo)
pMutacao=0.1
tamPopulacao=10
pais = 2

#Criando os Cromossomos
def cromossomo(menor,maior):
    return[np.random.randint(menor,maior) for i in range(tamCromossomo)]

#Criando a população inicial
def populacao():
    return[cromossomo(0,9) for i in range(tamPopulacao)]

#Função de Avaliação e total da aptidão com a Função Exponencial
def fitness(cromossomo):
    
    fitness = 0
    for i in range(len(cromossomo)):
        if(cromossomo [i] == modelo[i]):
            fitness += 1
    return fitness

def totalFitness():
    
    pAvaliacao = []
    
    for i in range(tamCromossomo):
        fitness = 0
        fitness += np.exp(i) 
        pAvaliacao.append(fitness)
        
    tf = sum(pAvaliacao)/np.exp(tamCromossomo)
    total = tf
    return total

#Seleção dos pais e Cruzamento
def selecao_e_cruzamento(population):
    scored = [(fitness(i), i) for i in population]
    scored = [i[1] for i in sorted(scored)]
    population = scored
    selected = scored[(len(scored) - pais):]
    for i in range(len(population) - pais):
        point = random.randint(1, tamCromossomo - 1)
        parent = random.sample(selected, 2)
        population[i][:point] = parent[0][:point]
        population[i][point:] = parent[1][point:]
    return population

#Função de Mutação
def mutacao(population):
    for i in range(len(population) - pais):
        if(random.random() <= pMutacao):
            point = random.randint(0, tamCromossomo - 1)
            new_value = random.randint(1, 9)
            while(new_value == population[i][point]):
                new_value = random.randint(1,9)
            population[i][point] = new_value
    return population

#Programa Principal do AG
population = populacao()
print("População: {}".format(population))
print("\n")
print("Total de Aptidão:")
print(totalFitness())
print("\n")
for i in range(100):
    population = selecao_e_cruzamento(population)
    population = mutacao(population)
print("População Final: {}".format(population))
```
